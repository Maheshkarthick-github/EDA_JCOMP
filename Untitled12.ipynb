Outlier detection

```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df['Age']))
threshold = 3
outliers = df[(z_scores > threshold)]
print(outliers)

Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]
print("\nOutliers (IQR method):")
print(outliers)
```

```python
cleaned_df = df.drop(outliers.index)
```

```python
print("Shape before outlier removal:", df.shape)
print("Shape after outlier removal:", cleaned_df.shape)
print(cleaned_df.describe())
```

Feature extraction

```python
keywords = ['good', 'bad', 'excellent', 'poor']
for keyword in keywords:
    df[f'Contains {keyword.capitalize()}'] = df['Review Text'].apply(lambda x: 1 if keyword in str(x).lower() else 0)
```

```python
print(df[['Review Text'] + [f'Contains {keyword.capitalize()}' for keyword in keywords]].head())
```

```python
df['Average Word Length'] = df.apply(lambda row: np.mean([len(word) for word in str(row['Review Text']).split()]), axis=1)
```

```python
print(df[['Review Text', 'Average Word Length']].head())
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans


features = ['Age', 'Positive Feedback Count']

df_normalized = (df[features] - df[features].mean()) / df[features].std()

kmeans = KMeans(n_clusters=3, random_state=0)
df['cluster'] = kmeans.fit_predict(df_normalized[features])

centroids = kmeans.cluster_centers_

df['distance'] = np.sqrt(((df_normalized[features] - centroids[df['cluster']])**2).sum(axis=1))

threshold = 2.0  # Adjust this threshold as needed
cleaned_df = df[df['distance'] <= threshold]

outliers = df[df['distance'] > threshold]
print("Outliers:")
print(outliers)

print("Cleaned DataFrame:")
print(cleaned_df)
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2, RFE
from sklearn.ensemble import RandomForestClassifier

data = {
    'Clothing ID': [1, 2, 3, 4, 5],
    'Age': [25, 30, 35, 40, 45],
    'Title': ['Great dress', 'Nice shirt', 'Comfortable pants', 'Lovely blouse', 'Stylish jacket'],
    'Review Text': ['I love this dress!', 'The shirt fits perfectly.', 'These pants are so comfy.', 'The blouse is beautiful.', 'The jacket is very stylish.'],
    'Rating': [5, 4, 5, 5, 4],
    'Positive Feedback Count': [10, 5, 8, 12, 6],
    'Division Name': ['General', 'General', 'General Petite', 'General', 'General Petite'],
    'Department Name': ['Dresses', 'Tops', 'Bottoms', 'Tops', 'Jackets'],
    'Class Name': ['Dresses', 'Blouses', 'Pants', 'Blouses', 'Jackets'],
    'Recommended IND': [1, 1, 1, 1, 0]  # 1 for recommended, 0 for not recommended
}

df = pd.DataFrame(data)

df_encoded = pd.get_dummies(df, columns=['Division Name', 'Department Name', 'Class Name'])

X = df_encoded.drop(['Clothing ID', 'Recommended IND', 'Title', 'Review Text'], axis=1)
y = df_encoded['Recommended IND']

selector = SelectKBest(score_func=chi2, k=3)
X_selected = selector.fit_transform(X, y)

print("Selected features using SelectKBest:", X.columns[selector.get_support()])

model = RandomForestClassifier()
rfe = RFE(model, n_features_to_select=3)
X_rfe = rfe.fit_transform(X, y)

print("Selected features using RFE:", X.columns[rfe.get_support()])
```
